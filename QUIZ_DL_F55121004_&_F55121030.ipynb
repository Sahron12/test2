# Memuat Data Tanpa Augmentasi Menggunakan Keras ImageDataGenerator

```python
from keras.preprocessing.image import ImageDataGenerator

# Membuat ImageDataGenerator tanpa augmentasi
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)  # rescale pixel values dan bagi dataset menjadi training dan validation set

# Memuat data training
train_generator = datagen.flow_from_directory(
    '/content/dataset_gambar',  # path dataset
    target_size=(224, 224),  # ukuran gambar yang diharapkan
    batch_size=32,
    class_mode='categorical',  # mode kelas untuk klasifikasi multi-kelas
    subset='training'  # subset training
)

# Memuat data validasi
validation_generator = datagen.flow_from_directory(
    '/content/dataset_gambar',  # path dataset
    target_size=(224, 224),  # ukuran gambar yang diharapkan
    batch_size=32,
    class_mode='categorical',  # mode kelas untuk klasifikasi multi-kelas
    subset='validation'  # subset validation
)
```

# Membuat Model VGG16 Menggunakan Keras Functional API dan Melakukan Training

```python
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense

# Membuat input layer
input_layer = Input(shape=(224, 224, 3))

# Blok 1
x = Conv2D(64, (3, 3), activation='relu', padding='same')(input_layer)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), strides=(2, 2))(x)

# Blok 2
x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), strides=(2, 2))(x)

# Blok 3
x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), strides=(2, 2))(x)

# Blok 4
x = Conv2D(512, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), strides=(2, 2))(x)

# Blok 5
x = Conv2D(512, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), strides=(2, 2))(x)

# Flatten layer
x = Flatten()(x)

# Fully connected layers
x = Dense(4096, activation='relu')(x)
x = Dense(4096, activation='relu')(x)
output_layer = Dense(3, activation='softmax')(x)  # Output layer dengan 3 kelas

# Membuat model
model = Model(inputs=input_layer, outputs=output_layer)

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Menampilkan ringkasan model
model.summary()

# Melatih model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size
)

# Evaluasi model pada data validasi
loss, accuracy = model.evaluate(validation_generator)
print(f'Validation Loss: {loss}, Validation Accuracy: {accuracy}')
```

# Melakukan Prediksi dan Menampilkan Hasilnya dalam Subplot 3x3

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image

# Path ke gambar-gambar yang ingin diprediksi
path_gambar = ["path/to/image1.jpg", "path/to/image2.jpg", ..., "path/to/image9.jpg"]

# Memuat gambar-gambar dan mengubahnya menjadi array numpy
images = []
for path in path_gambar:
    img = image.load_img(path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    images.append(img)

# Konkatenasi semua gambar menjadi satu array
images = np.concatenate(images, axis=0)

# Lakukan prediksi menggunakan model pada data gambar
predictions = model.predict(images)

# Mendapatkan label kelas hasil prediksi
predicted_classes = np.argmax(predictions, axis=1)

# Plot hasil prediksi menggunakan subplot ukuran 3x3
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(image.load_img(path_gambar[i]))
    plt.title(f'Predicted Class: {predicted_classes[i]}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Menyimpan model ke dalam file
model.save('/content/lokasi_penyimpanan/model_vgg16.h5')

# Menghapus model yang sudah ada dari memori (opsional)
del model

# Memuat model dari file yang sudah disimpan
from keras.models import load_model

loaded_model = load_model('/content/lokasi_penyimpanan/model_vgg16.h5')

# Menampilkan ringkasan model yang sudah dimuat
loaded_model.summary()
```
